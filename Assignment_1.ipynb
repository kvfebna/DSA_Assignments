Separate 3 classes into 3 dataframes

Class 2: Versicolor

Class 3: Virginica

# Scatter Plot of Iris Dataset

```python
plt.suptitle('Scatter Plot of Iris Dataset')
plt.subplot(1,2,1)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (PL)')
plt.ylabel('Petal Width (PW)')
plt.scatter(setosa['PL'],setosa['PW'],marker='^',c='r')
plt.scatter(versicolor['PL'],versicolor['PW'],marker='^',c='g')
plt.scatter(virginica['PL'],virginica['PW'],marker='^',c='b')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.subplot(1,2,2)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Sepal Width (SW)')
plt.scatter(setosa['SL'],setosa['SW'],marker='^',c='r')
plt.scatter(versicolor['SL'],versicolor['SW'],marker='^',c='g')
plt.scatter(virginica['SL'],virginica['SW'],marker='^',c='b')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.gcf().set_figwidth(10)
plt.subplots_adjust(wspace=.3)
```

OBSERVATION: 
1. From the first plot, it is clear that, as Petal Length increases Petal Width also increases for all the three clusters. 
2. We can differentiate Iris-Setosa cluster from the other two clusters in both of the plots. But Iris-Versicolor and Iris-Virginica clusters are overlapped.

# Box Plot of Iris Dataset 

```python
plt.suptitle('Box Plot of Iris Dataset')
plt.subplot(2,2,1) 
plt.title('Box Plot using SL')
sns.boxplot(x=iris['Classification'],y=iris['SL'])
plt.subplot(2,2,2)
sns.boxplot(x=iris['Classification'],y=iris['SW'])
plt.title('Box Plot using SW')
plt.subplot(2,2,3)
sns.boxplot(x=iris['Classification'],y=iris['PL'])
plt.title('Box Plot using PL')
plt.subplot(2,2,4)
sns.boxplot(x=iris['Classification'],y=iris['PW'])
plt.title('Box Plot using PW')
plt.gcf().set_figwidth(10)
plt.gcf().set_figheight(8)
plt.subplots_adjust(hspace=0.4)
plt.subplots_adjust(wspace=0.5)
```

OBSERVATION: 
1. Iris-Virginica class shows outliers in both boxplots using SL and SW. But there are no outliers in Iris-Setosa and Iris-Versicolor classes. 
2. In box plot using PL, outliers are present in both Iris-Setosa and Iris-Versicolor classes. Whereas only the class Iris-Setosa has outliers in box plot using PW.

# Heatmap of Iris Dataset

Remove object data type column from all the three dataframes

```python
setosa_new=setosa[['SL','SW','PL','PW']]
versicolor_new=versicolor[['SL','SW','PL','PW']]
virginica_new=virginica[['SL','SW','PL','PW']]
```

```python
plt.suptitle('Heatmap of Iris Dataset')
plt.subplot(1,3,1)
sns.heatmap(setosa_new.corr(),annot=True)
plt.title('Setosa')
plt.subplot(1,3,2)
sns.heatmap(versicolor_new.corr(),annot=True)
plt.title('Versicolor',pad=2)
plt.subplot(1,3,3)
sns.heatmap(virginica_new.corr(),annot=True)
plt.title('Virginica')
plt.gcf().set_figwidth(12)
plt.gcf().set_figheight(3.5)
plt.subplots_adjust(wspace=0.2)
```

OBSERVATION: In all the three classes, it shows a positive correlation between the all possible pairs of variables. i.e, As one variable increases in value, other one also increases.

# Histogram of Iris Dataset

```python
plt.suptitle('Histogram of Iris Dataset')
plt.subplot(4,3,1)
plt.hist(setosa['SL'])
plt.title('Setosa')
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Frequency')
plt.subplot(4,3,2)
plt.hist(versicolor['SL'])
plt.title('Versicolor')
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Frequency')
plt.subplot(4,3,3)
plt.hist(virginica['SL'])
plt.title('Virginica')
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Frequency')
plt.subplot(4,3,4)
plt.hist(setosa['SW'])
plt.title('Setosa')
plt.xlabel('Sepal Width (SW)')
plt.ylabel('Frequency')
plt.subplot(4,3,5)
plt.hist(versicolor['SW'])
plt.title('Versicolor')
plt.xlabel('Sepal Width (SW)')
plt.ylabel('Frequency')
plt.subplot(4,3,6)
plt.hist(virginica['SW'])
plt.title('Virginica')
plt.xlabel('Sepal Width (SW)')
plt.ylabel('Frequency')
plt.subplot(4,3,7)
plt.hist(setosa['PL'])
plt.title('Setosa')
plt.xlabel('Petal Length (PL)')
plt.ylabel('Frequency')
plt.subplot(4,3,8)
plt.hist(versicolor['PL'])
plt.title('Versicolor')
plt.xlabel('Petal Length (PL)')
plt.ylabel('Frequency')
plt.subplot(4,3,9)
plt.hist(virginica['PL'])
plt.title('Virginica')
plt.xlabel('Petal Length (PL)')
plt.ylabel('Frequency')
plt.subplot(4,3,10)
plt.hist(setosa['PW'])
plt.title('Setosa')
plt.xlabel('Petal Width (PW)')
plt.ylabel('Frequency')
plt.subplot(4,3,11)
plt.hist(versicolor['PW'])
plt.title('Versicolor')
plt.xlabel('Petal Width (PW)')
plt.ylabel('Frequency')
plt.subplot(4,3,12)
plt.hist(virginica['PW'])
plt.title('Virginica')
plt.xlabel('Petal Width (PW)')
plt.ylabel('Frequency')
plt.gcf().set_figwidth(11)
plt.gcf().set_figheight(11)
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=.7)
```

OBSERVATION:
1. The highest frequency of the Sepal Length is above 10 which is around 5 in Iris-Setosa class. Whereas in Iris-Versicolor and Iris-Virginica classes, highest frequency is below 10 and the values are around 5.6 and 6.9 respectively.
2. For Sepal Width, the highest frequency is around 15 for both Iris-Setosa and Iris-Virginica classes. For Iris-Versicolor class the high frequency is around 10.
3. In the case of Petal Length, Iris-Setosa class shows high frequency in the range 1.4 - 1.5. Whereas Iris-Versicolor class has a higher frequency around 4.5 and Iris-Virginica class has a higher frequency in the range 5 - 5.1.
4. Iris-Setosa class shows high frequency of greater than 20 for Petal Width which is above 0.2 and below 0.3. Also there are no values in between 0.3 and 0.4. And the values are absent in the range 1.3 - 1.4 for Iris-Versicolor class also.
